```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0, 20, 100)

beta = 5
f = lambda x: np.exp(-x) - np.exp(-x/beta)
f_grad = lambda x: -np.exp(-x) + np.exp(-x/beta)/beta
fig = plt.figure(figsize=(8,3.5))

# plot minimum
min_x = beta*np.log(beta)/(beta-1)

plt.plot(xx, f(xx), label='$V(r_{ij}) = e^{-r_{ij}} - e^{-r_{ij}/\\beta}$', c = "black", lw=1)
plt.plot(xx[xx<min_x], f(xx[xx<min_x]), c = "#C00000")
plt.plot(xx[xx>min_x], f(xx[xx>min_x]), c = "green")




plt.plot(min_x, f(min_x),'o', c = "midnightblue")
# make vertical line
plt.axvline(min_x, color='midnightblue', linestyle='--',  label='Minimum $r_{ij} = \\frac{\\beta \log(\\beta)}{\\beta-1}$',)
import seaborn as sns

repulsive = xx <= min_x +0.1
attractive = xx >= min_x

plt.fill_between(xx, -0.6, f(xx), where=repulsive, color='#C00000', alpha=0.3, edgecolor='none', label='Repulsive')
plt.fill_between(xx, 0., f(xx), where=repulsive, color='#C00000', alpha=0.2, edgecolor='none',)
plt.fill_between(xx, -0.6, f(xx), where=attractive, color='green', alpha=0.3, edgecolor='none', label='Attractive')
plt.fill_between(xx, 0, f(xx), where=attractive, color='green', alpha=0.2, edgecolor='none', )


# draw an arrow along the gradient at x = 1.5 and x = 3

sns.despine()
plt.legend()
plt.xlabel('$r_{ij}$')
plt.ylabel('$V(r_{ij})$')

plt.xlim(-1, 20)
plt.ylim(-0.6, 0.0)
```
