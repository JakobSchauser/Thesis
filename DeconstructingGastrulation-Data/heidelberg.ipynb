```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
with h5py.File('manual_annotations.h5') as f:
    for key in f.keys():
        print(key)

    table = f['table']
    print(table)
    table = np.array(table)
```

```python
x,y,z = table["RegionCenter_0"], table["RegionCenter_1"], table["RegionCenter_2"]

positions = np.array([x,y,z]).T
```

```python
from matplotlib.animation import FuncAnimation

# make 3d animation of the data

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')



def func(i):
    print(i)
    ax.clear()
    indx = table["timestep"] == i
    ax.scatter(positions[indx,0], positions[indx,2], -positions[indx,1], c='blue', marker='o')
    ax.set_title(f"timestep {i}")
    ax.set_xlim(0, 700)
    ax.set_ylim(-350, 350)
    ax.set_zlim(-350, 350)


ani = FuncAnimation(fig, func, frames=range(0, 100), interval=100)

ani.save('animation.mp4')
```
