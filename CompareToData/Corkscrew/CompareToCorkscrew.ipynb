```python
import numpy as np
import h5py 
import cv2
import matplotlib.pyplot as plt


def load_video(video_path):
    # load the video
    cap = cv2.VideoCapture(video_path)

    # save the frames
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)

    # convert the frames to a numpy array
    frames = np.array(frames)
    plt.imshow(frames[0])
    plt.show()
    return frames

sim = load_video('finale_nopmg.mp4')

data = load_video('../../StasVideos/corkscrew3.mp4')
```

```python
N = 4

fig, allaxs = plt.subplots(2, N, figsize= (10,7), sharex=True, sharey="row", constrained_layout=True)


videospeed = 20
simspeed = 200


for i in range(N):
    axs = allaxs[:, i]
    axs[0].imshow(sim[i * simspeed], aspect="auto")
    axs[1].imshow(data[i * videospeed], aspect="auto")

    # remove the x and y ticks

for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# fig.tight_layout()

plt.show()
```
