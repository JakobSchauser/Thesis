```python
import numpy as np
import h5py 
import cv2
import matplotlib.pyplot as plt


def load_video(video_path):
    # load the video
    cap = cv2.VideoCapture(video_path)

    # save the frames
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)

    # convert the frames to a numpy array
    frames = np.array(frames)
    plt.imshow(frames[0])
    plt.show()
    return frames

# sim = load_video('finale_nopmg.mp4')
```

```python
noventral = load_video('finale2_noventral.mp4')
nogb = load_video('finale3_nogb.mp4')
```

```python
def ind_to_m(ind):
    return ind/len(noventral)*18

def make_figure(data, timesteps):
    yn = 1
    if len(timesteps) > 3:
        yn = 2

    fig, axs = plt.subplots(yn, len(timesteps) // (yn ), sharey=True, figsize=(12, 12), constrained_layout=True)

    axs = axs.flatten()

    for i, t in enumerate(timesteps):
        axs[i].imshow(data[t])
        axs[i].set_title(f"t =  {ind_to_m(t):.0f} min", fontweight='bold', fontsize=20)
        axs[i].set_xticks([])
        axs[i].set_yticks([])

    # top
    s = 250
    d = 340
    yy = -90
    axs[0].text(yy, s, "Top view\n (dorsal)", rotation=90, fontsize=12, fontweight='bold',) 
    axs[0].text(yy, s+d, "Side view\n (lateral)", rotation=90, fontsize=12, fontweight='bold',) 
    axs[0].text(yy, s+2*d, "Bottom view\n   (ventral)", rotation=90, fontsize=12, fontweight='bold',) 

    plt.show()



```

```python
make_figure(noventral, [50, 100, 200])
```

```python
make_figure(nogb, [100, 300, 400, 550])
```
