```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0, 20, 100)

beta = 5
f = lambda x: np.exp(-x) - np.exp(-x/beta)

fig = plt.figure(figsize=(6,3))

plt.plot(xx, f(xx), label='$V(r_{ij}) = e^{-r_{ij}} - e^{-r_{ij}/\\beta}$', c = "teal")

# plot minimum
min_x = beta*np.log(beta)/(beta-1)


plt.plot(min_x, f(min_x),'o', label='Minimum $r_{ij} = \\frac{\\beta \log(\\beta)}{\\beta-1}$', c = "#901A1E")
import seaborn as sns
sns.despine()
plt.legend()
plt.xlabel('$r_{ij}$')
plt.ylabel('$V(r_{ij})$')
```
