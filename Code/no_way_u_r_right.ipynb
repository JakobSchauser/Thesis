```python
import numpy as np
```

```python
N = int(1e8)
random_uniform_2d = np.random.uniform(0, 1, (N, 2))

# make a 2d histogram using numpy
bins, xedges, yedges = np.histogram2d(random_uniform_2d[:,0], random_uniform_2d[:,1], bins=50)



# plot the density of the points
import matplotlib.pyplot as plt
plt.imshow(bins, extent=(xedges[0], xedges[-1], yedges[0], yedges[-1]), origin='lower')
```
