```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0, 20, 100)

beta = 5
f = lambda x: np.exp(-x) - np.exp(-x/beta)

plt.plot(xx, f(xx), label='V(x) = $e^{-x} - e^{-x/\\beta}$', c = "teal")

# plot minimum
min_x = beta*np.log(beta)/(beta-1)

plt.plot(min_x, f(min_x),'o', label='Minimum $x = \\frac{\\beta \log(\\beta)}{\\beta-1}$', c = "#901A1E")
import seaborn as sns
sns.despine()
plt.legend()
plt.xlabel('x')
plt.ylabel('V(x)')
```
